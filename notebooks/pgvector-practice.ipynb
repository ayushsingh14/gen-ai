## Create the schema and insert sample documents

This uses the helper code in `src/pgvector_practice/db.py`.

```python
from pgvector_practice.db import create_schema, insert_document, vector_search

create_schema()
print('Schema created.')

# Example document insert. Replace embedding with your own vector values.
sample_embedding = [0.1] * 1536
insert_document('Notebook sample document', sample_embedding)
print('Sample document inserted.')
```

## Run a vector search

Use a query embedding to find the nearest documents.

```python
query_embedding = [0.1] * 1536
results = vector_search(query_embedding, limit=5)
results
```

## Notes

- `pgvector` embeddings typically come from an encoder such as OpenAI, Sentence Transformers, or another embedding model.
- Make sure your embeddings match the vector dimension defined in the schema (`1536`).
- If you want to try document storage and retrieval at scale, add more rows to the `documents` table and experiment with `ORDER BY embedding <-> query_embedding`.
